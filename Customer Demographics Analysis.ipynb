```python
## Import Libaries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load datasets 
data = pd.read_csv("./Cost_prediction.csv")
# Display the first few rows
data.head()
```

```python
# Check for missing values
print(data.isnull().sum())

# Optionally, drop rows with missing data or fill with appropriate values
# data = data.dropna()  # Drop rows with missing values
# data['gender'] = data['gender'].fillna('Unknown')  # Example: filling missing gender values
```

```python
# Now, let’s analyze sales by gender.
# Group by gender and calculate the sum of store sales and unit sales
gender_sales = data.groupby('gender')[['store_sales(in millions)', 'unit_sales(in millions)']].sum()
print(gender_sales )
```

```python

# Plot sales by gender
plt.figure(figsize=(8, 5))
sns.barplot(x=gender_sales.index, y=gender_sales['store_sales(in millions)'], palette='viridis')
plt.title('Total Sales by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Sales (in millions)')
plt.show()

# Optionally, analyze unit sales as well
plt.figure(figsize=(8, 5))
sns.barplot(x=gender_sales.index, y=gender_sales['unit_sales(in millions)'], palette='viridis')
plt.title('Total Unit Sales by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Unit Sales (in millions)')
plt.show()
```

```python
# Next, let’s see how education levels are related to sales.
# Group by education level and calculate the sum of store sales and unit sales
education_sales = data.groupby('education')[['store_sales(in millions)', 'unit_sales(in millions)']].sum()
print(education_sales)

# Plot sales by education level
plt.figure(figsize=(10, 6))
sns.barplot(x=education_sales.index, y=education_sales['store_sales(in millions)'], palette='coolwarm')
plt.title('Total Sales by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Total Sales (in millions)')
plt.xticks(rotation=45)
plt.show()

# Analyze unit sales as well
plt.figure(figsize=(10, 6))
sns.barplot(x=education_sales.index, y=education_sales['unit_sales(in millions)'], palette='coolwarm')
plt.title('Total Unit Sales by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Total Unit Sales (in millions)')
plt.xticks(rotation=45)
plt.show()
```

```python
# You can also explore how sales_country affects sales.
# Group by country and calculate the sum of store sales and unit sales
country_sales = data.groupby('sales_country')[['store_sales(in millions)', 'unit_sales(in millions)']].sum()
print(country_sales)

# Plot sales by country
plt.figure(figsize=(12, 6))
sns.barplot(x=country_sales.index, y=country_sales['store_sales(in millions)'], palette='crest')
plt.title('Total Sales by Country')
plt.xlabel('Country')
plt.ylabel('Total Sales (in millions)')
plt.xticks(rotation=90)
plt.show()

# Analyze unit sales as well
plt.figure(figsize=(12, 6))
sns.barplot(x=country_sales.index, y=country_sales['unit_sales(in millions)'], palette='crest')
plt.title('Total Unit Sales by Country')
plt.xlabel('Country')
plt.ylabel('Total Unit Sales (in millions)')
plt.xticks(rotation=90)
plt.show()
```

To test whether demographic variables significantly impact sales, you can perform a One-Way ANOVA or t-test.

For Gender vs Sales:

```python
from scipy.stats import ttest_ind
# Separate sales data by gender
male_sales = data[data['gender'] == 'M']['store_sales(in millions)']
female_sales = data[data['gender'] == 'F']['store_sales(in millions)']

# Perform t-test
t_stat, p_value = ttest_ind(male_sales, female_sales)

print(f'T-statistic: {t_stat}, P-value: {p_value}')
```

**Interpretation** **of** **p-value** **(0.9739)** 

A p-value of **0.9739** indicates that there is no statistically significant difference in sales between genders in this dataset. Here's a breakdown of what that means:

Interpreting the p-value:

- Null Hypothesis (H₀): There is no difference in sales between genders (male and female).

- Alternative Hypothesis (H₁): There is a significant difference in sales between genders.

Since the p-value (0.9739) is much greater than the commonly used significance level (α = 0.05), we fail to reject the null hypothesis. This means there is **no** **statistical** **evidence** to suggest that gender affects sales in this dataset.






```python
# Perform ANOVA for education levels and sales
from scipy.stats import f_oneway

# Group sales by education level
groups = [data[data['education'] == edu]['store_sales(in millions)'] for edu in data['education'].unique()]

# Perform ANOVA
f_stat, p_value = f_oneway(*groups)

print(f'F-statistic: {f_stat}, P-value: {p_value}')
```

A p-value of **0.0027** in the context of an ANOVA test comparing **education** **level** **vs.** **sales** indicates the following:

**Hypotheses**

- Null Hypothesis(H₀):There is no difference in sales across different educational levels. 

- Alternative Hypothesis (H₁): There is a significant difference in sales across educational levels.

**Statistical** **Significance** **:** Since the p-value **(0.0027)** is less than the typical significance level of **0.05**, the result is statistically significant. This provides strong evidence to **reject** **the** **null** **hypothesis**, which states that there is no difference in sales across different education levels.

**Interpretation** **:** In practical terms, this result suggests that **education** **level** **has** **a** **meaningful** **effect** **on** **sales**. That is, the average sales amounts vary significantly depending on customers’ education levels.

**Next** **Steps** **:** 

Since I was able to establish that education level affects sales, I took the following steps:

- Examined the mean sales for each education level to identify which specific groups spend more or less.
- Performed a post-hoc tests (Using Tukey's HSD test) to determine which education groups differ significantly.This provided more granular insight into which levels of education are driving the observed differences.

Examine Mean Sales for Each Education Level
This will help us understand the average sales amount for each educational group


```python
# Calculate mean sales for each education level
education_mean_sales = data.groupby('education')['store_sales(in millions)'].mean()
print("Mean Sales by Education Level:")
print(education_mean_sales)

# Visualize mean sales by education level

plt.figure(figsize=(10, 6))
sns.barplot(x=education_mean_sales.index, y=education_mean_sales.values, palette='coolwarm')
plt.title('Mean Sales by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Mean Sales (in millions)')
plt.xticks(rotation=45)
plt.show()
```

Based on the mean sales values for each education group, we can interpret the following:

High School Degree holders have the highest mean sales at approximately 6.61 million.
Bachelors Degree and Partial High School holders have close mean sales, around 6.51–6.54 million.
Graduate Degree holders have slightly lower mean sales compared to other groups, at about 6.39 million.


Given that the average sales are close across groups, the Tukey’s HSD test results will be especially helpful in confirming if any of these observed differences are statistically significant. If any education level pairs show a significant difference in sales, this could indicate that specific education levels have unique purchasing behaviors or tendencies.

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Extract the relevant columns for the test
education_sales = data[['education', 'store_sales(in millions)']]

# Perform Tukey's HSD test
tukey_results = pairwise_tukeyhsd(endog=education_sales['store_sales(in millions)'], 
                                  groups=education_sales['education'], 
                                  alpha=0.05)

# Print the results
print(tukey_results)

# Plot the results
tukey_results.plot_simultaneous(figsize=(10, 6))
plt.title("Tukey's HSD Test Results for Education vs. Sales")
plt.xlabel('Mean Difference in Sales (in millions)')
plt.show()
```

**Key** **Findings** **:**

- **Significant** **Difference** **Between** **Graduate** **and** **High** **School** **Degree** **Holders**
The only pair showing a statistically significant difference in mean sales is between **Graduate** **Degree** and **High** **School** **Degree** **holders**. This is supported by a mean difference (meandiff) of **0.2224** and an adjusted p-value (p-adj) of **0.0084**, which is below the 0.05 significance threshold. The null hypothesis of no difference in means is rejected for this pair, indicating that these two groups have statistically different average sales.

- **No** **Significant** **Differences** **Among** **Other** **Groups**
All other group comparisons have adjusted p-values above 0.05 and a rejection status of False, meaning there is no evidence to suggest significant differences in sales between these education groups.

**Interpretation**

- **Graduate** **Degree** **vs.** **High** **School** **Degree** **:**
Customers with a High School degree tend to have higher average sales than those with Graduate degrees, as indicated by the positive mean difference.

- **Practical** **Insight** **:**
Although the difference is statistically significant, the absolute value (0.22 million) is relatively small. Still, this may reflect behavioral differences where customers with a High School education spend more on average than those with Graduate degrees.
